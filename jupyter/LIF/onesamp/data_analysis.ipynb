```python
import h5py
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
out_deg = h5py.File('out_deg.h5', 'r')
print([key for key in out_deg.keys()])
```

```python
key = 'p_rew_0_start_1_dq_1'
df_out_deg = pd.read_hdf('out_deg.h5',key)
df_in_deg = pd.read_hdf('in_deg.h5',key)
# df.shape[0] #this is the number of rows
bins_number = 101
count_out_deg = np.zeros((bins_number,df_out_deg.shape[1]))
count_in_deg = np.zeros((bins_number,df_in_deg.shape[1]))
for i in range(df_out_deg.shape[1]) :
    count, division_in_deg = np.histogram(df_in_deg.loc[:,i], bins = bins_number,range =(0,101))
    count_in_deg[:,i] = count
    count, division_out_deg = np.histogram(df_out_deg.loc[:,i], bins = bins_number,range =(0,101))
    count_out_deg[:,i] = count

x = np.arange(0,100,0.05)
y = np.arange(0,101,1)
X,Y = np.meshgrid(x,y)
plt.scatter(X,Y,c = count_out_deg)
plt.colorbar()
plt.show()
```
